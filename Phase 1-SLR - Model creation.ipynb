```python
import pandas                        #Pandas is a python library which is used to analyse datas in the form of tabulation. It
import pandas as pd                  #    is an open-source library that provides high-performance data manipulation in Python.
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Salary_Data.csv") #csv is a comma seperated value files which we have as input and output data in the form
                                         #of tabulation. Here this csv file is read and assigned to the variable dataset 
```

```python
dataset                                # checking whether the datas are read properly and assigned to that variable
```

```python
#input/output split  -here input and output datas are splitted and assigned to two different variables
```

```python
independent=dataset[["YearsExperience"]]
dependent=dataset[["Salary"]]
```

```python
independent
```

```python
dependent
```

```python
plt.scatter(independent,dependent)
plt.xlabel('Years of Experience',fontsize=20)
plt.ylabel('Salary',fontsize=20)
plt.show
```

```python
#Split train and test set - here the splitted input and output dataset are further splitted to test and training dataset
#The sklearn library contains a lot of efficient tools for machine learning and statistical modeling including 
#                                classification, regression, clustering and dimensionality reduction.
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
x_train
```

```python
y_train
```

```python
#Model creation/Learning phase - In this phase model is created by giving the input and output values of training set using
#                                                                                                      linear regression.
#In this case when regressor.fit is performed within the coding the weight and bias will be calculated.Thus AI brain is created.
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train, y_train)
```

```python
#Displaying calculated weight and bias
```

```python
Weight=regressor.coef_                #Weight of the model is calculated using the code regressor.coef_
Weight
```

```python
Bias=regressor.intercept_             #Bias of the model is calculated using the code regressor.intercept_
Bias
```

```python
#Evaluation Metrics
```

```python
y_pred=regressor.predict(x_test)     #Using regressor.predict, the predicted output(Dependent variable) for the 
y_pred                               #      input values(Independent  variable) of test set is predicted here.
```

```python
from sklearn.metrics import r2_score  #using the validation parameter r2_score , actual output and the predicted output is 
r_score=r2_score(y_test, y_pred)      #                                            compared and the score is calclated. 
```

```python
r_score                              #If the score is near to '1' it is a good model 
                                     #                        '0' it is a bad model
                                     #In this model the value is near to '1' and so it is a good model.
```

```python
#To save model
```

```python
#Pickle is used for serializing and de-serializing Python object structures, also called marshalling or flattening. 
#Serialization refers to the process of converting an object in memory to a byte stream that can be stored on disk or sent
#                                                                                                         over a network.
```

```python
import pickle
filename="Finalized_model_linear.sav"
```

```python
#The dump() method of the pickle module in Python, converts a Python object hierarchy into a byte stream
#"wb" mode opens the file in binary format for writing.
```

```python
pickle.dump(regressor, open(filename,'wb')) 
```

```python
#Loading the saved model
#The primary argument of pickle load function is the file object that you get by opening the file in read-binary (rb) mode.
```

```python
loaded_model=pickle.load(open("Finalized_model_linear.sav",'rb'))
result=loaded_model.predict([[13]])
```
